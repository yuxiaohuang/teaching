## Handling missing data

### Imputing missing values
The code below shows how to use the mean of a numerical feature to impute its missing values.

```python
from sklearn.impute import SimpleImputer

# If there are missing values
if len(df_miss['var']) > 0:
    # The SimpleImputer
    si = SimpleImputer(missing_values=np.nan, strategy='mean')

    # Impute the variables with missing values in df_train, df_val and df_test 
    df_train[df_miss['var']] = si.fit_transform(df_train[df_miss['var']])
    df_val[df_miss['var']] = si.transform(df_val[df_miss['var']])
    df_test[df_miss['var']] = si.transform(df_test[df_miss['var']])
```

## Surface and contour plot of MSE

The code below shows how to draw the surface plot of training MSE of the simple linear regression (with feature LotFrontage and target SalePrice).

```python
# Create a figure
fig = plt.figure(figsize=(8, 5))

# Create the axes
ax = plt.axes(projection='3d', proj_type='ortho')

# Get w0 and w1
w0, w1 = np.linspace(-5, 5, 100), np.linspace(-5, 5, 100)

# Get the x and y axes (w0 and w1)
x, y = np.meshgrid(w0, w1)

# Get the prediction on the training data
y_train_pred = np.dot(x_LotFrontage_train, y.reshape(1, -1)) + x.reshape(1, -1)

# Get the z axis (MSE)
z = np.sum((y_train_pred - np.array(y_train).reshape(-1, 1)) ** 2, axis=0) / x_LotFrontage_train.shape[0]

# The surface plot between x, y and z
ax.plot_surface(x, y, z.reshape(x.shape), cmap=cm.coolwarm)

# Set x-axis
ax.set_xlabel('$b$', labelpad=10)

# Set y-axis
ax.set_ylabel('$w$', labelpad=10)

# Set z-axis
ax.set_zlabel('$L(b, w)$', labelpad=5)

# Save and show the figure
plt.tight_layout()
plt.savefig(abspath_curr + 'result/figure/obj_LotFrontage_SalePrice_surface.pdf')
plt.show()
```
